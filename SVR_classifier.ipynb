```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000, n_features=2,n_classes = 2, n_clusters_per_class=2, n_redundant=0, random_state=1)
```

```python
x
```

```python
y
```

```python
pd.DataFrame(x)
```

```python
pd.DataFrame(x)[0]
```

```python
pd.DataFrame(x)[1]
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
# SVM

from sklearn.svm import SVC
```

```python
clf = SVC(kernel = 'linear')
```

```python
clf
```

```python
clf.fit(x_train,y_train)
```

```python
clf.coef_
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
# Hyperparameter tuning

from sklearn.model_selection import GridSearchCV

params = {'C': [0.1, 0.2, 1, 2, 3, 10, 50, 100],
         'gamma': [1, 0.1, 0.2, 0.001, 0.003],
         'kernel': ['linear']}
```

```python
grid = GridSearchCV(SVC(),param_grid = params,cv=5,verbose=3)
```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.best_estimator_.predict(x_test)
```

```python
y_pred
```

```python
print(accuracy_score(y_test,y_pred))
```
