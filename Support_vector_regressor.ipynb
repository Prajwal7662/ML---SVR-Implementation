```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_samples = 1000,n_features=2,n_targets=1,noise=3)
```

```python
x
```

```python
y
```

```python
pd.DataFrame(x)[0]
```

```python
pd.DataFrame(x)[1]
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue = y)
```

```python
# train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel='linear')
```

```python
svr
```

```python
svr.fit(x_train,y_train)
```

```python
svr.coef_
```

```python
y_pred = svr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
 r2_score(y_test,y_pred)
```

```python
# hyperparameter tuning

from sklearn.model_selection import GridSearchCV

params = {'C': [0.1, 0.2, 1, 2, 3, 10, 50, 100],
         'gamma': [1, 0.1, 0.2, 0.001, 0.003],
         'kernel': ['linear'],
          'epsilon' :[0.01,0.1,0.2,0.3]}
```

```python
 grid = GridSearchCV(SVR(),param_grid=params,cv=5,verbose=3)
```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
final_model = grid.best_estimator_
```

```python
y_pred = final_model.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
